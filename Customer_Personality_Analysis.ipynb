```python
# Import pandas (you only need to do this once per session, but good practice!)
import pandas as pd

# 1. Load the data using the EXACT filename and the tab separator
df = pd.read_csv('/marketing_campaign (2).csv', sep='\t')

# 2. Check the "shape" of the data (how many rows and columns we have)
print(f"Dataset size: {df.shape[0]} rows and {df.shape[1]} columns.\n")

# 3. Look at the basic info: data types and missing values
df.info()

# 4. Display the first 5 rows
df.head()
```

```python
# 1. Drop the rows where 'Income' is missing
df = df.dropna(subset=['Income'])

# 2. Convert 'Dt_Customer' from text to an actual datetime format
# Note: We use format='%d-%m-%Y' because the data is formatted Day-Month-Year
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

# 3. Check our work to make sure it worked!
print(f"New dataset size: {df.shape[0]} rows (We successfully dropped 24 rows!)")
print(f"Missing Income values remaining: {df['Income'].isna().sum()}")
print(f"Dt_Customer data type is now: {df['Dt_Customer'].dtype}")
```

```python
# 1. Calculate Age (Assuming dataset's present year is 2015)
df['Age'] = 2015 - df['Year_Birth']

# 2. Calculate Total Spent on all items
df['Total_Spend'] = (df['MntWines'] + df['MntFruits'] +
                     df['MntMeatProducts'] + df['MntFishProducts'] +
                     df['MntSweetProducts'] + df['MntGoldProds'])

# 3. Calculate Total Children
df['Total_Children'] = df['Kidhome'] + df['Teenhome']

# 4. Calculate Customer Tenure (in days)
# We find the most recent enrollment date to act as our "present day"
latest_date = df['Dt_Customer'].max()
df['Days_as_Customer'] = (latest_date - df['Dt_Customer']).dt.days

# 5. Drop the old columns we don't need anymore to keep our dataset tidy
# Note: Z_CostContact and Z_Revenue are useless columns in this dataset where every row has the same value.
columns_to_drop = ['Year_Birth', 'Dt_Customer', 'Kidhome', 'Teenhome', 'Z_CostContact', 'Z_Revenue']
df = df.drop(columns=columns_to_drop)

# Peek at our awesome new engineered columns!
df[['Age', 'Total_Spend', 'Total_Children', 'Days_as_Customer']].head()
```
